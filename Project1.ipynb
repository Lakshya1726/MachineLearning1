```python
deaths.drop(['Name','Allegiances','Death Year','Book of Death','Death Chapter','Book Intro Chapter'],axis=1,inplace=True)
```

```python
deaths.head()
```

```python
deaths.fillna(0)
```

```python
deaths.drop('SoS',axis=1).head()
```

```python
deaths['SoS'].head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(deaths.drop('SoS',axis=1), 
                                                    deaths['SoS'], test_size=0.30, 
                                                    random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
prediction=logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy=confusion_matrix(y_test,prediction)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,prediction)
accuracy
```

```python
prediction
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction))
```
